```python
# Dependenciesd
import pandas as pd
```

```python
# read file into a dataframe

vine_df = pd.read_csv("vine_df.csv")
vine_df
```

```python
# Step 1
vine_df_votes = vine_df.loc[vine_df["total_votes"] >= 20]
vine_df_votes
```

```python
# Step 2

vine_df_50 = vine_df_votes.loc[(vine_df_votes["helpful_votes"]/vine_df_votes["total_votes"]) >= 0.50]
vine_df_50
```

```python
# Step 3

paid_reviews = vine_df_50.loc[vine_df_50["vine"] == 'Y']

paid_reviews
```

```python
# Step 4
unpaid_reviews = vine_df_50.loc[vine_df_50["vine"] == 'N']

unpaid_reviews
```

```python
# Step 5
# paid 5 star

paid_5star = paid_reviews.loc[paid_reviews["star_rating"] == 5]["star_rating"].count()
paid_5star
```

```python
# total paid
total_paid = paid_reviews["star_rating"].count()
total_paid
```

```python
# unpaid 5 stars
unpaid_5star = unpaid_reviews.loc[unpaid_reviews["star_rating"] == 5]["star_rating"].count()
unpaid_5star
```

```python
total_unpaid = unpaid_reviews["star_rating"].count()
total_unpaid
```
